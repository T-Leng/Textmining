```python
def tokenize_and_extract_bigrams(text):
    tokens = word_tokenize(text)
    bigrams = [' '.join(bg) for bg in everygrams(tokens, 2, 2)]
    return bigrams
```

```python
def display_topics(model, feature_names, num_top_bigrams):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx + 1))
        top_bigrams = []
        for i in topic.argsort()[:-num_top_bigrams - 1:-1]:
            bigram_words = feature_names[i].split()
            if len(bigram_words) == 2:
                aspect = bigram_words[0]
                opinion = bigram_words[1]
                top_bigrams.append((aspect, opinion))
            if len(top_bigrams) == num_top_bigrams:
                break
        for aspect, opinion in top_bigrams:
            print(aspect, opinion)
        print()
```
