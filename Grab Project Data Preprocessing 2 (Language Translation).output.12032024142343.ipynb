```python
import pandas as pd
from deep_translator import GoogleTranslator
import time
```

```python
df = pd.read_csv("GRAB_REVIEWS_CLEANED.csv")
df
```

```python
translator = GoogleTranslator(source='auto', target='en')
```

```python
def translate_to_english(text):
    try:
        translated_text = translator.translate(text)
        return translated_text
    except Exception as e:
        print(f"Translation failed for text: {text}. Error: {e}")
        return None  # Return the original text if translation fails
    
start_time = time.time()
df.loc[100000:, 'translated_review'] = df['review_text'].iloc[100000:].apply(translate_to_english)

end_time = time.time()

total_time = end_time - start_time
print("Total time taken:", total_time, "seconds")
```

```python
df.to_csv('GRAB_REVIEWS_CLEANED.csv', index=False)
```
