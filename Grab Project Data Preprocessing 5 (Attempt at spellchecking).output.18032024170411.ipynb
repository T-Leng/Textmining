```python
from symspellpy import SymSpell, Verbosity
import pkg_resources
import pandas as pd
```

```python
df = pd.read_csv("GRAB_REVIEWS_CLEANED.csv")
df
```

```python
df.dropna(subset=['translated_review'], inplace=True)
df.dropna(subset=['translated_review_cleaned'], inplace=True)
```

```python
symspell = SymSpell(max_dictionary_edit_distance=2)
bigram_path = pkg_resources.resource_filename("symspellpy", "frequency_bigramdictionary_en_243_342.txt")
symspell.load_dictionary(bigram_path, term_index=0, count_index=2)

df['translated_review_corrected'] = ""

for i, sentence in enumerate(df['translated_review_cleaned']):
    suggestions = symspell.lookup_compound(sentence, max_edit_distance=2)
    df.at[i, 'translated_review_corrected'] = suggestions[0].term
```

```python
df.to_csv('GRAB_REVIEWS_CLEANED.csv', index=False)
```
